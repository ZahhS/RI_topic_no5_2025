```python
import cbor
import pickle
# Charger le fichier CBOR
with open('train.pages.cbor', 'rb') as f:
    data_cbor = cbor.load(f)

# Sauvegarder les données en format Pickle
with open('train.pages.cbor.pkl', 'wb') as f:
    pickle.dump(data_cbor, f)
```

```python
import pandas as pd

# Charger le fichier CBOR
with open('train.pages.cbor', 'rb') as f:
    data_cbor = cbor.load(f)

print(data_cbor)

# Convertir les données en DataFrame
df = pd.DataFrame(data_cbor)

# Sauvegarder en format TSV
df.to_csv('train.pages.cbor.tsv', sep='\t', index=False)
```
